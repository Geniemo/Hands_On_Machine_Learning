# 여러 타임 스텝 앞을 예측하기
첫 번째 방법은 이미 훈련된 모델을 사용하여 다음 값을 예측한 다음<br>
이 값을 입력으로 추가하는 것입니다(예측된 값이 실제 발견된 값인 것처럼 다룹니다.).

다음 코드와 같이 구현할 수 있습니다.

```python
series = generate_time_series(1, n_steps + 10)
X_new, Y_new = series[:, :n_steps], series[:, n_steps:]
X = X_new
for step_ahead in range(10):
    y_pred_one = model.predict(X[:, step_ahead:])[:, np.newaxis, :]
    X = np.concatenate([X, y_pred_one], axis=1)

Y_pred = X[:, n_steps:]
```

```python
def plot_multiple_forecasts(X, Y, Y_pred):
    n_steps = X.shape[1]
    ahead = Y.shape[1]
    plot_series(X[0, :, 0])
    plt.plot(np.arange(n_steps, n_steps + ahead), Y[0, :, 0], "bo-", label="Actual")
    plt.plot(np.arange(n_steps, n_steps + ahead), Y_pred[0, :, 0], "rx-", label="Forecast", markersize=10)
    plt.axis([0, n_steps + ahead, -1, 1])
    plt.legend(fontsize=14)

plot_multiple_forecasts(X_new, Y_new, Y_pred)
plt.show()
```

위 그래프에서 볼 수 있듯이 가까운 미래에 대한 예측이 더 정확합니다.<br>
미래의 타임 스텝은 오차가 누적될 수 있기 때문입니다.

이 방식을 검증 세트에 적용하면 약 0.029의 MSE가 나오는데,<br>
이전 모델보다 크게 높지만 훨씬 어려운 작업을 했으므로 단순히 비교하기는 어렵습니다.<br>
따라서 같은 작업을 하는 단순한 예측이나 간단한 선형 모델과 비교하는 게 의미가 있습니다.

이렇게 비교를 한다면 이 모델이 단순한 예측이나 간단한 선형 모델보다 월등히 앞서는 것을 알 수 있습니다.

두 번째 방법은 RNN을 훈련해 다음 값 10개를 한 번에 예측하는 것입니다.<br>
시퀀스-투-벡터 모델을 사용하고 1개가 아니라 값 10개를 출력합니다.

먼저 타깃을 다음 10개의 값이 담긴 벡터로 바꾸겠습니다.

```python
n_steps = 50
series = generate_time_series(10000, n_steps + 10)
X_train, Y_train = series[:7000, :n_steps], series[:7000, -10:, 0]
X_valid, Y_valid = series[7000:9000, :n_steps], series[7000:9000, -10:, 0]
X_test, Y_test = series[9000:, :n_steps], series[9000:, -10:, 0]
```

```python
model = keras.models.Sequential([
    keras.layers.SimpleRNN(20, return_sequences=True, input_shape=[None, 1]),
    keras.layers.SimpleRNN(20),
    keras.layers.Dense(10)
])

model.compile(loss="mse", optimizer="adam")
history = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

이제 예측을 한 번 해보겠습니다.

```python
series = generate_time_series(1, 50 + 10)
X_new, Y_new = series[:, :50, :], series[:, -10:, :]
Y_pred = model.predict(X_new)[..., np.newaxis]
```

```python
plot_multiple_forecasts(X_new, Y_new, Y_pred)
plt.show()
```

이 모델은 잘 작동하고, 다음 10개 타임 스텝에 대한 MSE는 약 0.008입니다.
선형 모델보다 훨씬 좋지만, 아직 개선할 여지가 있습니다.

마지막 타임 스텝에서만 다음 값 10개를 예측하도록 훈련하는 대신<br>
모든 타임 스텝에서 다음 값 10개를 예측하도록 모델을 훈련할 수 있습니다.

다르게 말하면 이 시퀀스-투-벡터 RNN을 시퀀스-투-시퀀스 RNN으로 바꾸는 것입니다.

타깃 시퀀스를 준비해봅시다.

```python
n_steps = 50
series = generate_time_series(10000, n_steps + 10)
X_train = series[:7000, :n_steps]
X_valid = series[7000:9000, :n_steps]
X_test = series[9000:, :n_steps]
Y = np.empty((10000, n_steps, 10))  # 각 타깃은 10D 벡터의 시퀀스
for step_ahead in range(1, 10 + 1):
    Y[..., step_ahead - 1] = series[..., step_ahead:step_ahead + n_steps, 0]
Y_train = Y[:7000]
Y_valid = Y[7000:9000]
Y_test = Y[9000:]
```

이 모델을 시퀀스-투-시퀀스 모델로 바꾸려면 마지막 층을 포함한 모든 순환 층에 return_sequences=True를 지정해줘야 합니다. 그 다음 모든 타임 스텝에서 출력을 Dense층에 적용해야 합니다. 이를 위해서는 keras의 TimeDistributed층을 사용하면 됩니다. 이 층은 다른 층을 감싸서 입력 시퀀스의 모든 타임 스텝에 이를 적용합니다. 각 타임 스텝을 별개의 샘플처럼 다루게 입력의 크기를 바꿔 이를 효과적으로 수행합니다.

```python
model = keras.models.Sequential([
    keras.layers.SimpleRNN(20, return_sequences=True, input_shape=[None, 1]),
    keras.layers.SimpleRNN(20, return_sequences=True),
    keras.layers.TimeDistributed(keras.layers.Dense(10))
])
```

평가를 위해 마지막 타임 스텝의 출력에 대한 MSE만을 계산하는 사용자 정의 지표를 사용하겠습니다.

```python
def last_time_step_mse(Y_true, Y_pred):
    return keras.metrics.mean_squared_error(Y_true[:, -1], Y_pred[:, -1])

model.compile(loss="mse", optimizer=keras.optimizers.Adam(learning_rate=0.01), metrics=[last_time_step_mse])
history = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

검증 MSE로 0.006을 얻었습니다. 이전 결과보다 매우 좋은 결과입니다.

```python
series = generate_time_series(1, 50 + 10)
X_new, Y_new = series[:, :50, :], series[:, 50:, :]
Y_pred = model.predict(X_new)[:, -1][..., np.newaxis]
```

```python
plot_multiple_forecasts(X_new, Y_new, Y_pred)
plt.show()
```

이런 방식을 처음 모델과 결합할 수 있습니다.<br>
이 RNN으로 다음 10개를 예측하고 이 값을 입력 시계열에 연결하고,<br>
모델을 다시 사용해 다음 값 10개를 예측합니다.
