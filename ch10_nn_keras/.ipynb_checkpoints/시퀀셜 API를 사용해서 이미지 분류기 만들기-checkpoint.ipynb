# 데이터셋 적재

케라스는 여러 데이터셋을 다운로드하고 적재할 수 있는 유틸리티 함수를 제공합니다.<br>
패션 MNIST 데이터셋을 적재해보겠습니다.

```python
import tensorflow as tf
from tensorflow import keras
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

훈련 세트는 6만개가 할당됩니다.

```python
X_train_full.shape
```

각 픽셀의 강도는 바이트(0 ~ 255)로 표현됩니다.

```python
X_train_full.dtype
```

전체 훈련 세트를 검증 세트와 조금 더 작은 훈련 세트로 나눠보겠습니다.<br>
또한 픽셀 강도를 0 ~ 1 범위의 실수로 바꾸겠습니다.

```python
X_valid, X_train = X_train_full[:5000] / 255., X_train_full[5000:] / 255.
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
X_test = X_test / 255.
```

MNIST는 레이블이 5면 이미지가 손글씨 숫자 5를 나타낸다는 것을 알 수 있었지만,<br>
패션 MNIST는 레이블에 해당하는 아이템을 나타내기 위하여 클래스 이름의 리스트를 만들어야 합니다.

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```
