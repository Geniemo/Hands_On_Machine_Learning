```python
import tensorflow as tf
from tensorflow import keras
```

아래와 같은 방법으로 ResNet-50 모델을 불러올 수 있습니다.

```python
model = keras.applications.resnet50.ResNet50(weights='imagenet')
```

이 코드는 ResNet-50 모델을 만들고 이미지넷 데이터셋에서 사전훈련된 가중치를 다운로드합니다.

이 모델을 사용할 때에는 이미지가 적절한 크기인지 확인해야 합니다.

ResNet-50 모델은 224 \\(\times\\) 224 픽셀 크기의 이미지를 기대합니다.<br>
텐서플로의 tf.image.resize() 함수로 이미지의 크기를 바꿀 수 있습니다.<br>
(이 함수는 가로세로 비율을 유지하지 않으므로 가로세로 비율을 유지하려면 미리 이미지를 알맞게 자르거나 tf.image.crop_and_resize() 함수를 사용하면 됩니다.)

```python
from sklearn.datasets import load_sample_image
import numpy as np
china = load_sample_image('china.jpg') / 255
flower = load_sample_image('flower.jpg') / 255
images = np.array([china, flower])
```

```python
import matplotlib.pyplot as plt

images_resized = tf.image.resize(images, [224, 224])
plt.imshow(images_resized[0])
```

사전훈련된 모델은 이미지가 적절한 방식으로 전처리되었다고 생각합니다.<br>
그리고 어떤 입력을 기대할지는 모델에 따라 다를 수 있습니다.

이를 위해 모델마다 이미지를 전처리해주는 preprocess_input() 함수를 제공합니다.<br>
이 함수는 픽셀값이 0에서 255 사이라고 가정합니다.

따라서 images_resized에 255를 곱해야 합니다.

```python
inputs = keras.applications.resnet50.preprocess_input(images_resized * 255)
```

이제 사전훈련된 모델을 사용해 예측을 수행할 수 있습니다.

```python
Y_proba = model.predict(inputs)
Y_proba
```

predict의 결과는 통상적인 구조대로 행이 하나의 이미지이고 열이 하나의 클래스인 행렬입니다. 최상위 K개 예측의 클래스 이름과 예측 클래스의 추정 확률을 출력하려면 decode_predictions() 함수를 사용합니다.

```python
top_K = keras.applications.resnet.decode_predictions(Y_proba, top=5)
```

```python
for image_index in range(len(images)):
    print(f'이미지 #{image_index}')
    for class_id, name, y_proba in top_K[image_index]:
        print('  {} - {:12s} {:.2f}%'.format(class_id, name, y_proba * 100))
```

두 이미지의 정답 클래스(monastery, daisy)가 상위 결과 5개에 포함되어 있습니다.<br>
모델이 1000개의 클래스 중에서 찾는다는 것을 고려하면 꽤 괜찮은 결과입니다.
