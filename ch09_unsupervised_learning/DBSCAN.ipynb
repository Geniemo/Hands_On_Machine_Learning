sklearn의 반달 모양 데이터셋을 이용하겠습니다.

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=1000, noise=0.05, random_state=42)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

모든 샘플의 레이블은 인스턴스 변수 labels\_에 저장되어 있습니다.

```python
dbscan.labels_[:10]
```

일부 샘플의 인덱스는 -1입니다.<br>
알고리즘이 이 샘플을 이상치로 판단했다는 의미입니다.

핵심 샘플의 인덱스는 인스턴스 변수 core\_sample\_indices\_에 저장되어 있습니다.

```python
len(dbscan.core_sample_indices_)
```

```python
dbscan.core_sample_indices_[:10]
```

핵심 샘플 자체는 인스턴스 변수 components\_에 저장되어 있습니다.

```python
dbscan.components_
```

이 군집 결과는 아래 그림과 같습니다.

```python
import matplotlib.pyplot as plt
import os

image = plt.imread(os.path.join('..', 'images', 'unsupervised_learning', 'dbscan_plot.png'))
plt.figure(figsize=(16, 8))
plt.imshow(image)
plt.show()
```

그림에서 보듯이 클러스터를 7개나 만들었고 많은 샘플들을 이상치로 판단했습니다.

eps를 0.2로 증가해서 이웃 범위를 넓히면 오른쪽처럼 완벽한 군집을 얻습니다.

DBSCAN 클래스는 predict() 메소드를 제공하지 않고 fit_predict() 메소드를 제공합니다.

다시 말해서 이 알고리즘은 새로운 샘플에 대해 클러스터를 예측할 수 없습니다.<br>
이런 구현 겾렁은 다른 분류 알고리즘이 이런 작업을 더 잘 수행할 수 있기 때문입니다.

따라서 사용자가 필요한 예측기를 선택해야 합니다.

예를 들어 KNeighborsClassifier를 훈련해보겠습니다.

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_])
```

```python
import numpy as np
X_new = np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

이 결정 경계를 시각화하면 아래와 같습니다.

```python
image = plt.imread(os.path.join('..', 'images', 'unsupervised_learning', 'cluster_classification_plot.png'))
plt.figure(figsize=(16, 8))
plt.imshow(image)
plt.show()
```

훈련 세트에 이상치가 없으므로 클러스터가 멀리 떨어져있더라도 분류기는 항상 클러스터 하나를 선택합니다.<br>
최대 거리를 사용하면 두 클러스터에서 멀리 떨어진 샘플을 이상치로 간단히 분류할 수 있습니다.<br>
KNeighborsClassifier의 kneighbors() 메소드를 사용합니다.<br>
이 메소드에 샘플을 전달하면 훈련 세트에서 가장 가까운 k개 이웃의 거리와 인덱스를 반환합니다.<br>
(k개 열을 가진 행렬 두 개를 반환합니다.)

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors=1)
y_pred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx]
y_pred[y_dist > 0.2] = -1
y_pred.ravel()
```

DBSCAN은 간단하지만 강력합니다.<br>
DBSCAN은 클러스터의 모양과 개수에 상관없이 감지할 수 있는 능력이 있습니다.

이상치에 안정적이고 하이퍼파라미터가 두개뿐입니다.(eps, min\_samples)
